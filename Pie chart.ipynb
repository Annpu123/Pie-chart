# pie

```python
x= [45,65,63,75]
y=['Republican','Democrate','Orthodox','Conservate']
ex=[0.1,0.0,0.0,0.0,]
plt.pie(x,labels=y, explode=ex, autopct="%0.2f%%", shadow=True)
plt.show()
```
